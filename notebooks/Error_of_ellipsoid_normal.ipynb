```python
from spice import vsep
```

```python
from mars_spice import MarsSpicer
```

```python
mspice = MarsSpicer()
```

```python
lats = range(90)
```

```python
seps = []
```

```python
for lat in lats:    mspice.set_spoint_by(lat=lat,lon=0)    seps.append(degrees(vsep(mspice.snormal,mspice.spoint)))
```

```python
plot(lats,seps)
```
