```python
from osgeo import gdal
from lxml import etree
from pymars import pdstools
reload(pdstools)
from pykml.factory import KML_ElementMaker as KML
from pykml.factory import GX_ElementMaker as GX
```

```python
label = pdstools.PDSLabel('/Users/maye/data/hirise/inca/ESP_022607_0985_RED.LBL')
```

```python
stylename = "sn_ylw-pushpin"
stylename2 = "msn_ylw-pushpin"
stylename3 = "sh_ylw-pushpin"
doc = KML.kml(
    KML.Document(
        KML.Name("Test Position"),
        KML.Style(
            KML.IconStyle(
                KML.scale(1.1),
                KML.Icon(
                    KML.href("http://maps.google.com/mapfiles/kml/pushpin/ylw-pushpin.png")
                ),
            ),
            KML.LineStyle(
                KML.color('ffff0000'),
                KML.width('3')
            ),
            KML.PolyStyle(
                KML.color('800080ff'),
            ),
            id=stylename
        ),
        KML.StyleMap(
            KML.Pair(
                KML.key('normal'),
                KML.styleUrl('#{0}'.format(stylename)),
            ),
            KML.Pair(
                KML.key('highlight'),
                KML.styleUrl('#{0}'.format(stylename3)),
            ),
            id=stylename2
        ),
        KML.Style(
            KML.IconStyle(
                KML.scale(1.3),
                KML.Icon(
                    KML.href("http://maps.google.com/mapfiles/kml/pushpin/ylw-pushpin.png")
                ),
            ),
            KML.LineStyle(
                KML.color('ffff0000'),
                KML.width(3)
            ),
            KML.PolyStyle(
                KML.color('800080ff')
            ),
            id=stylename3
        ),
    )
)
```

```python
pm_poly=  KML.Placemark(
       KML.name("Some Inca"),
       KML.LookAt(
         KML.longitude(str(label.meanlon-360)),
         KML.latitude(str(label.meanlat)),
         KML.altitude('0'),
         KML.heading('0'),
         KML.tilt('0'),
         KML.range('88600'),
         GX.altitudeMode('relativeToSeaFloor')
       ),
       KML.styleUrl('#{0}'.format(stylename2)),
       KML.Polygon(
        KML.tessellate('1'),
        KML.outerBoundaryIs(
          KML.LinearRing(
             KML.coordinates(
              "{maxlat},{westmost}\n{minlat},{westmost}\n"
              "{minlat},{eastmost}\n{maxlat},{eastmost}\n"
              "{maxlat},{westmost}".format(westmost=label.maxlat,
                                           maxlat=label.westmost-360,
                                           eastmost=label.minlat,
                                           minlat=label.eastmost-360)
             ),
          ),
        ),
    ),
)
```

```python
doc.Document.append(pm_poly)
```

```python
outfile = open('testring.kml','w')
outfile.write(etree.tostring(doc, pretty_print=True))
outfile.close()
```
