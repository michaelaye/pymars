#### Save time for later

```python
start_utc = mspice.utc
```

#### Setting coords to lon=0, lat=-40

```python
mspice.set_spoint_by(lon=0, lat=-40)
```

#### Check how flux on flat ground and on (so far) untilted element compares (should be same). Units are W/m**2. The martian solar constant depends on the above set time.

```python
print(mspice.F_flat,mspice.F_tilt)
```

#### Now tilt local surface element by 33 degrees (taken average from RSL Science paper).  (default aspect is towards north)

```python
mspice.tilt = 33
```

#### Compare again flat ground flux again with flux on tilted element

```python
print(mspice.F_flat, mspice.F_tilt)
```

#### Check illumination angles. (Phase and Emission are set to 0 here, because I am using the MarsSpicer class in pure geometry mode, without having an observer. This has the advantage of using the default SPICE kernel for huge time intervals. If we would have set an observer, phase and emission angle would be calculated as well, but then the time needs to right to fit within the currently loaded SPICE kernels. Eventually this will get an interface to provide the kernels required.

```python
mspice.illum_angles()
```

#### What is the current local solar time at these coordinates?

```python
mspice.local_soltime
```

#### Now the loops over time. Per step jump ahead one hour, process in total 40*120 steps to reach approx. southern autumn start.

```python
times, energies = mspice.time_series('F_aspect',3600,no_of_steps=40*120,provide_times='l_s')
```

```python
mspice.utc = start_utc
```

#### Now turn the tilted surface element to east by 90 degrees and recalculate the irradiation for the same time range again.

```python
mspice.aspect=90
```

```python
energies_90 = mspice.time_series('F_aspect',3600, no_of_steps=40*120)
```

#### The same for 180 and 270 degree aspect turn.

```python
mspice.utc=start_utcmspice.aspect = 180energies_180 = mspice.time_series('F_aspect',3600, no_of_steps=40*120)
```

```python
mspice.utc=start_utcmspice.aspect = 270energies_270 = mspice.time_series('F_aspect', 3600, no_of_steps=40*120)
```

```python
plot(times,energies,times,energies_90,times,energies_180, times,energies_270,alpha=0.5)legend(('energies','_90','_180','_270'),loc=0)xlabel('L_s [deg]')ylabel('W/(m^2 * 3600s)') 
```
